```python
# Import librairies
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
import os


from itertools import cycle
from plotly.subplots import make_subplots

path = os.getcwd() + '/'
```

```python
# Immport data
df = pd.read_csv(path + 'Speed+Dating+Data.csv', encoding='windows-1252')
```

```python
# Create dics to improve readbility by mapping numbers and their corresponding string from the codebook.
labels_race = {2: "European/Caucasian-American", 
4: "Asian/Pacific Islander/Asian-American", 
6 : "Other", 3 : "Latino/Hispanic American", 
1 : "Black/African American"}
labels_field = {1: "Law", 
2: "Math", 
3 : "Social Science, Psychologist" ,
4 :  "Medical Science, Pharmaceuticals, and Bio Tech", 
5 : "Engineering",  
6 : "English/Creative Writing/ Journalism", 
7 : "History/Religion/Philosophy",
8 : "Business/Econ/Finance",
9 : "Education, Academia",
10 : "Biological Sciences/Chemistry/Physics",
11 : "Social Work", 
12 : "Undergrad/undecided",
13 : "Political Science/International Affairs", 
14 : "Film",
15 : "Fine Arts/Arts Administration",
16 : "Languages",
17 : "Architecture",
18 : "Other"
}
```

```python
# Create two dataframes for males and females
df_male = df[df['gender'] == 1]
df_female = df[df['gender'] == 0]
# Map values from dictionnairies
df_male.race = df_male.race.map(labels_race)
df_female.race = df_female.race.map(labels_race)
df_male.field_cd = df_male.field_cd.map(labels_field)
df_female.field_cd = df_female.field_cd.map(labels_field)
```

```python
# Visulalise number of match for all participants
fig_match = px.bar(df.groupby('match').count(), y = 'id', labels={'match': 'Match', 'id':'Count'}, title="Number of matched participant")
fig_match.show()
```

```python
# Create age distribution for males and females
fig_age_male = px.bar(df_male.age.value_counts(), y=df_male.age.value_counts(), labels={'y' : 'Count', 'index':'age'}, title="Age distribution for male population")
fig_age_female = px.bar(df_female.age.value_counts(), y=df_female.age.value_counts(), labels={'y' : 'Count', 'index':'age'}, title="Age distribution for female population")
```

```python
# Create subplot layout to better visluaise both sexes
fig = make_subplots(rows=1, cols=2, subplot_titles=('Age distribution for male population', 'Age distribution for female population'))
fig.add_trace(fig_age_male['data'][0], row=1,col=1)
fig.add_trace(fig_age_female['data'][0], row=1,col=2)
fig.show()
```

```python
# Creare and view pie charts to view distribution of race and field by sex.
fig_race_male = px.pie(df_male, "race", title="Race of males" )
fig_race_female = px.pie(df_female, 'race', title="Race of females")

fig_field_male = px.pie(df_male, 'field_cd', title="Fields of male population")
fig_field_female = px.pie(df_female, 'field_cd', title="Fields of female population")

fig_race_male.show()
fig_field_male.show()
fig_race_female.show()
fig_field_female.show()
```

```python
# View declared importance of each attribute for males and demales
fig_pref_male = px.line(df_male[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']].sum(), y= df_male[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']].sum(), labels={"y" : "Sum value for given attribute", "index" : "attribute"}, title="Importance of each attribute for male population")
fig_pref_female = px.line(df_female[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']].sum(), y= df_female[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']].sum(), labels={"y" : "Sum value for given attribute", "index" : "attribute"}, title="Importance of each attribute for female population")
fig_pref_male.show()
fig_pref_female.show()
```

```python
# Isolate for both sexs does that matched
match_df_male = df_male[df_male['match'] == 1]
match_df_female = df_female[df_female['match'] == 1]
```

```python
fig_grading_male = px.line(match_df_male[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].sum(), y= match_df_male[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].sum(), labels={"y" : "Sum value for given attribute", "index" : "attribute"}, title="Rating of partner for male population")
fig_grading_female = px.line(match_df_female[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].sum(), y= df_female[['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].sum(), labels={"y" : "Sum value for given attribute", "index" : "attribute"}, title="Rating of partner for female population")
fig_grading_male.show()
fig_grading_female.show()
```

```python
# Create correlation matricies to understand the gap between 
corr_male = match_df_male[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].corr()
corr_female = match_df_female[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1', 'attr', 'sinc', 'intel', 'fun', 'amb', 'shar']].corr()
```
